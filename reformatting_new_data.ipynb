```python
%matplotlib inline
from __future__ import division
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy.random as npr
import re
```

```python
newdat = pd.read_excel('new_hair_shaft_only_from_alison.xlsx', 0)
```

```python
newdat_long = pd.wide_to_long(newdat, 'Hair_', i=['individual_id_old', 'Position'], j='idx')
```

```python
newdat_long.reset_index(inplace=True)
```

```python
newdat_long.drop(labels=['Major_Allele', 'Minor_Allele'], inplace=True, axis=1)
```

```python
newdat_long.rename(columns={'Hair_': 'hair', 'MAF_blood':'blood', 'MAF_cheek':'cheek', 'individual_id_old':'individual_id', 'Position':'position'}, inplace=True)
```

```python
newdat_long.dropna(subset=['hair'], inplace=True)
```
